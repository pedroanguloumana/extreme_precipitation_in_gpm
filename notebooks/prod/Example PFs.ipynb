```python
region = cio_region()
pf_stats = load_pf_stats(region)
size = pf_stats.num_pixels * 25
max_pr = pf_stats.max_precip
loc = pf_stats.max_5mmhr_cluster_precip/pf_stats.total_precip
```

```python
idx = (max_pr>=100) & (0.7<loc)&(loc<0.95) & (size>1e4)
stats = pf_stats.where(idx).dropna()
file = stats.pf_filename.iloc[0]
print(file)
```

```python
idx = (max_pr>=100) & (0.2<loc)&(loc<0.5) & (size>1e4)
stats = pf_stats.where(idx).dropna()
file = stats.pf_filename.iloc[1]
print(file)
```
